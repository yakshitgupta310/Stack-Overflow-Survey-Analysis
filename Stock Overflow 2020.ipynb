```python
#looking at df.info we get that we need to change data types of few columns
df.YearsCodePro.replace(["Less than 1 year","More than 50 years"],["0.5","51"],inplace=True)
df.YearsCode.replace(["Less than 1 year","More than 50 years"],["0.5","51"],inplace=True)
change={"YearsCode":float , "YearsCodePro":float, "Age":int}
df=df.astype(change,errors='ignore')
```

```python
df.set_index("Respondent",inplace=True)
```

```python
df
```

```python
dflang=df[~df["LanguageWorkedWith"].isna()] 
```

```python
languages={}
for i in dflang["LanguageWorkedWith"].str.split(";"):
    for j in i:
        if j in languages:
            languages[j]+=1
        else:
            languages[j]=1
```

```python
languages   #Dictionary with count of how many people have worked with a certain Language
```

```python
from heapq import nlargest
N=5
topLang=nlargest(N,languages,key=languages.get)
topLang #List of Top Languages
```

```python
Langper={}
for i in topLang:
    Langper[i]=f"{languages[i]/df.shape[0]:.2f}"
    
Langper  #shows percentage of Top lang 
```
